```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import seaborn as sns


N=1000

x = np.zeros(N)
y = np.zeros(N)
z = np.zeros(N)
    
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
plt.grid(alpha=0.15)
        
for i in range(1,N):
        
    R = np.random.randint(0,6) #to choose which direction X,Y,Y
    #d = np.random.randint(1,4) #to choose a random distance in that random direction
        
    if R == 0:
        x[i] = x[i-1] - 1
        y[i] = y[i-1]
        z[i] = z[i-1]
    elif R==1:
        x[i] = x[i-1] + 1
        y[i] = y[i-1]
        z[i] = z[i-1]
    elif R==2:
        x[i] = x[i-1]
        y[i] = y[i-1] - 1
        z[i] = z[i-1]
    elif R==3:
        x[i] = x[i-1]
        y[i] = y[i-1] + 1
        z[i] = z[i-1]
    elif R==4:
        x[i] = x[i-1]
        y[i] = y[i-1]
        z[i] = z[i-1] - 1
    elif R==5:
        x[i] = x[i-1]
        y[i] = y[i-1]
        z[i] = z[i-1] + 1
        
    x_p = [x[i-1],x[i]]
    y_p = [y[i-1],y[i]]
    z_p = [z[i-1],z[i]]
        
    plt.plot(x_p,y_p,z_p,'-', 
        color='#4CB391',
        alpha =0.6,
        linewidth=0.8)

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.view_init(15,-45)
    

coord = list(zip(x,y,z))
df = pd.DataFrame(coord,columns=['X','Y','Z'])
sns.jointplot(x='X',y='Y',data=df,
    kind="hex",
    color="#4CB391",
    gridsize=15)
sns.set_theme(style="ticks")
plt.grid(alpha=0.2)


plt.show()
```
