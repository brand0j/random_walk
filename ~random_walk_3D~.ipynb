```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import seaborn as sns
import time

start_time = time.time()

N=1000
x = np.zeros(N)
y = np.zeros(N)
z = np.zeros(N)
    
fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection='3d')
        
for i in range(1,N):
        
    R = np.random.randint(0,6)
        
    if R == 0:
        x[i] = x[i-1] -1
        y[i] = y[i-1]
        z[i] = z[i-1]
    elif R==1:
        x[i] = x[i-1] +1
        y[i] = y[i-1]
        z[i] = z[i-1]
    elif R==2:
        x[i] = x[i-1]
        y[i] = y[i-1] -1
        z[i] = z[i-1]
    elif R==3:
        x[i] = x[i-1]
        y[i] = y[i-1] +1
        z[i] = z[i-1]
    elif R==4:
        x[i] = x[i-1]
        y[i] = y[i-1]
        z[i] = z[i-1] -1
    elif R==5:
        x[i] = x[i-1]
        y[i] = y[i-1]
        z[i] = z[i-1] +1
        
    x_p = [x[i-1],x[i]]
    y_p = [y[i-1],y[i]]
    z_p = [z[i-1],z[i]]
        
    plt.plot(x_p,y_p,z_p,'-o', color='blue',alpha =0.7,linewidth=0.75,markersize=0.4)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.view_init(15,-60)
    
coord = list(zip(x,y,z))
plt.show()

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
df = pd.DataFrame(coord,columns=['X','Y','Z'])
sns.jointplot(x='X',y='Y',data=df,kind='hex')
```
