```python
## IMPORT ##
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure

# A random walk with n number of steps
def random_walk(n): 
    
    # Starts at (0,0), initialize for replacement
    x=np.zeros(n+1) 
    y=np.zeros(n+1)

    # Set up our plot (plt.grid() makes this hard to see. Remove right & top spines)
    fig = plt.figure(figsize=(15,15))
    ax=fig.add_subplot()
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    

    # Generate a random n*pi/2 value where n is a random integer
    # This makes it so either we move in the x-direction OR the y-direction, but not both
    for i in range(1,n+1):
    
        # x,y values
        theta = (np.random.randint(1,5)*np.pi)/2
        x[i] = x[i-1]+ int(np.cos(theta))
        y[i] = y[i-1]+ int(np.sin(theta))
        
        # Need this to draw the lines from point to point
        x_values = [x[i-1],x[i]]
        y_values = [y[i-1],y[i]]
        
        # Styling the lines. -- dash with markers for the points. 
        # Set alpha <1 so we can visualize paths that are taken more than once
        plt.plot(x_values,y_values,'--o',color='blue',alpha=0.3,linewidth=0.5,markersize=1)
    
    # Plot starting & ending points so they are clearly visible
    plt.plot(x[0],y[0],'o',color='#00EE76',markersize=7,label='START') #start point(green)
    plt.plot(x[n],y[n],'o',color='red',markersize=7,label='END') #end point(red)
    plt.legend()
    plt.title('Random walk with n= '+str(n)+' steps')
    
    # for smaller n: plot the coordinates of each point
    if n <= 150:
        for x,y in zip(x,y):
            # Creating labels for our points along the random walk
            label= f"({int(x)},{int(y)})"
            plt.annotate(label,(x,y),textcoords="offset points",xytext=(5,10),ha='left')
            
    plt.show()
```

```python
# Call random_walk(n) and specify the amount of n-steps you want to take # 
random_walk(1000) 
```
